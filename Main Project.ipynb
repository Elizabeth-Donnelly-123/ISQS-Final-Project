```python
from IPython.display import Audio
import numpy as np

sample_rate = 44100
notes = {'E4': 329.63, 'D4': 293.66, 'C4': 261.63, 'G4': 392.00}
melody = [
    ('E4', 0.4), ('D4', 0.4), ('C4', 0.4), ('D4', 0.4),
    ('E4', 0.4), ('E4', 0.4), ('E4', 0.6),
    ('D4', 0.4), ('D4', 0.4), ('D4', 0.6),
    ('E4', 0.4), ('G4', 0.4), ('G4', 0.6),
    ('E4', 0.4), ('D4', 0.4), ('C4', 0.4), ('D4', 0.4),
    ('E4', 0.4), ('E4', 0.4), ('E4', 0.4), ('E4', 0.4),
    ('D4', 0.4), ('D4', 0.4), ('E4', 0.4), ('D4', 0.4),
    ('C4', 0.8),
]

song = []
for note, duration in melody:
    t = np.linspace(0, duration, int(sample_rate * duration), False)
    tone = np.sin(notes[note] * t * 2 * np.pi)
    song.extend(tone)

Audio(song, rate=sample_rate, autoplay=True)
```
